```python
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("../Data_csv/crop_yield.csv")
df
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.neighbors import KNeighborsRegressor
```

```python
le = LabelEncoder()
df['Season'] = le.fit_transform(df['Season'])
df['State'] = le.fit_transform(df['State'])
df['Crop'] = le.fit_transform(df['Crop'])
```

```python
df
```

```python
# Taking X as feature and Y as Target
X = df[['Season', 'State', 'Crop', 'Annual_Rainfall', 'Fertilizer', 'Pesticide']]
Y = df['Yield']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# to standarize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# KNN
from sklearn.metrics import accuracy_score
KNN = KNeighborsRegressor()

KNN.fit(X_train_scaled,y_train)
y_pred = KNN.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
cv_scores = cross_val_score(KNN, X_train_scaled, y_train, cv=5, scoring='r2')

print(" Results:")
print(f"RMSE: {rmse:.2f}")
print(f"R2 Score: {r2:.2f}")
print(f"Cross-validation R2 Score: {cv_scores.mean():.2f} (+/- {cv_scores.std()*2:.2f})")
```

```python
import pickle
# Dump the trained Naive Bayes classifier with Pickle
KNN_pkl_filename = '../utils/KNN.pkl'
# Open the file to save as pkl file
KNN_Model_pkl = open(KNN_pkl_filename,'wb')
pickle.dump(KNN, KNN_Model_pkl)
# Close the pickle instances
KNN_Model_pkl.close()
```
